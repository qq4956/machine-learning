## 梯度下降法
梯度下降法不是机器学习算法,是一个机遇搜索的最优化函数.

### 原理 
由于梯度存在方向,yan'zhe


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1,6,141)
```

```python
plot_x
```

```python
#绘制损失函数
plot_y = (plot_x-2.5)**2 -1
```

```python
plt.plot(plot_x,plot_y)
plt.show()
```

```python
# 损失函数的导数
def dJ(theta):
    return 2*(theta - 2.5)
```

```python
# 损失函数
def J(theta):
    return (theta - 2.5)**2 -1
```

```python
#eta为学习率
eta = 0.01
#theta从0开始取
theta = 0.0
#两个J的差值小于精度要求,我们认定导数为0
epsilon = 1e-8

while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    
    if(abs(J(theta) - J(last_theta))<epsilon):
        break

print (theta)
print (J(theta))
print (gradient)
```

```python
## 可视化梯度下降的过程


eta = 0.1
theta = 0.0
epsilon = 1e-8
theta_history = []

while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    
    if(abs(J(theta) - J(last_theta))<epsilon):
        break
        
plt.plot(plot_x,J(plot_x))
plt.plot(np.array(theta_history),J(np.array(theta_history)),marker='+',color = 'r')
plt.show()
```

```python
红线为梯度下降的过程.
刚开始移动大,是因为导数比较陡
```

```python
len(theta_history)
```

```python
## 当eta取值过大的后果,J过大..为inf


eta = 1.1
theta = 0.0
epsilon = 1e-8
theta_history = []
times = 10
i = 0

while i < times:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    i = i+1
    
    
    if(abs(J(theta) - J(last_theta))<epsilon):
        break
        
plt.plot(plot_x,J(plot_x))
plt.plot(np.array(theta_history),J(np.array(theta_history)),marker='+',color = 'r')
plt.show()
```

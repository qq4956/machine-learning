## 多分类



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg1 = LogisticRegression()
log_reg1.fit(X_train,y_train)
```

默认采用multi_class='ovr' 进行分类

```python
def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap)
```

```python
plot_decision_boundary(log_reg1, axis=[4, 8.5, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.scatter(X[y==2,0], X[y==2,1])
plt.show()
```

采用ovr方式分类的结果.

```python
log_reg2 = LogisticRegression(multi_class='multinomial',solver = 'newton-cg')
log_reg2.fit(X_train,y_train)
```

```python
plot_decision_boundary(log_reg2, axis=[4, 8.5, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.scatter(X[y==2,0], X[y==2,1])
plt.show()
```

```python
log_reg1.score(X_test,y_test)
```

```python
log_reg2.score(X_test,y_test)
```

采用ovo的方式,比ovr的方式,精确度更高.因为是仅对两个分类进行投票.

### 采用所有的数据

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
log_reg1 = LogisticRegression()
log_reg1.fit(X_train,y_train)
log_reg1.score(X_test,y_test)
```

```python
log_reg2 = LogisticRegression(multi_class='multinomial',solver = 'newton-cg')
log_reg2.fit(X_train,y_train)
log_reg2.score(X_test,y_test)
```

采用ovo准确率很高..

### OvO OvR

```python
from sklearn.multiclass import OneVsOneClassifier
OvO = OneVsOneClassifier(log_reg1)
```

```python
OvO.fit(X_train,y_train)
```

```python
OvO.score(X_test,y_test)
```

```python
from sklearn.multiclass import OneVsRestClassifier
OvR = OneVsRestClassifier(log_reg1)
OvR.fit(X_train,y_train)
OvR.score(X_test,y_test)
```

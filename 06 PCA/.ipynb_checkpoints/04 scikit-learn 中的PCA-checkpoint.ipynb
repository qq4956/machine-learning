```python
import numpy as np
import matplotlib.pyplot as plt
from  sklearn import datasets
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

## 不使用PCA,直接KNN全量数据,测试效率和精度

```python
%%time
## 使用KNN 全量数据fit,测试时间
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train,y_train)
```

```python
knn_clf.score(X_test,y_test)
```

### 使用PCA,降低维度到2维,测试维度和精度

```python
from sklearn.decomposition import PCA
pca= PCA(2)
pca.fit(X_train,y_train)
```

```python
X_train_reduction = pca.transform(X_train)
X_test_reduction = pca.transform(X_test)
```

```python
%%time
knn_clf_pca_2 =  KNeighborsClassifier()
knn_clf_pca_2.fit(X_train_reduction,y_train)
```

```python
knn_clf_pca_2.score(X_test_reduction,y_test)
```

使用2维数据,精度不够.

```python
X_train_reduction.shape
```

## 使用PCA找到合适的维度 explained_variance_ratio

```python
pca = PCA(X.shape[1])
```

explained_variance_ratio_ : 降序排列,各个维度所占的方差百分比.

```python
pca.fit(X_train)
pca.explained_variance_ratio_
```

```python
plt.plot([i for i in range(X.shape[1])],
         [np.sum(pca.explained_variance_ratio_[:i+1]) for i in range(X.shape[1])])
plt.show()
```

横坐标为"主成分",纵坐标为"主成分所占的方差"累计百分比.

```python
#"主成分所占的方差"累计百分比为95%
pca = PCA(0.95)
pca.fit(X_train)
```

```python
pca.n_components_
#28个数据所占的方差"累计百分比为95%
```

```python
X_train_reduction = pca.transform(X_train)
X_test_reduction = pca.transform(X_test)
```

```python
%%time
knn_clf =  KNeighborsClassifier()
knn_clf.fit(X_train_reduction,y_train)
```

```python
knn_clf.score(X_test_reduction,y_test)
```

精度降低,速度提升

### PCA二维的作用:可视化

```python
pca = PCA(2)
pca.fit(X)
X_reduction = pca.transform(X)
```

```python
for i in range(len(y)):
    plt.scatter(X_reduction[y == i,0],X_reduction[y == i,1],alpha = 0.5)
plt.show()
```

可见蓝色与橙色,可以考虑用2个维度区分

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

### 

```python
boston = datasets.load_boston()
```

```python
x = boston.data[:,5] #boston
```

```python
x.shape
```

```python
y = boston.target
```

```python
plt.scatter(x,y,alpha = 0.5)
plt.show()
```

```python
x = x[y<50]
y = y[y<50]
#fancy index!!!!! xy
plt.scatter(x,y,alpha = 0.5)
plt.show()
```

```python
#,MSE(),RMSE(),MAE().
```

```python
#
```

```python
#split
from mymodule.model_selection import  train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,seed = 666)
```

```python
#
from mymodule.SimpleLinearRegression import SimpleLinearRegression
```

```python
reg = SimpleLinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.a_
```

```python
reg.b_
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,reg.predict(x_train),c= 'r')
plt.show()
```

```python
#y_hat
y_predict = reg.predict(x_test)
```

### MSE

```python
#np.sum,.
mse_test = np.sum((y_predict -y_test)**2)/len(y_test)
mse_test
```

## RMSE

```python
rmse_test = np.sqrt(np.sum((y_predict - y_test)**2)/len(y_test))
rmse_test
```

## MAE

```python
mae_test = np.sum(np.abs(y_predict - y_test))/len(y_test)
mae_test
```

## scikit-learn MSE MAE

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mean_squared_error(y_test,y_predict)
```

```python
mean_absolute_error(y_test,y_predict)
```

## R

```python
1 - mean_squared_error(y_test,y_predict)/ np.var(y_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_predict)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x  = np.linspace(0,10,100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(y)
```

```python
#绘制两条直线
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = ':')
plt.show()
```

```python
#查看指定数据
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = ':')
plt.xlim(-5,15)
plt.ylim(0,1.5)
plt.show()
```

```python
#绘制两条直线
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = ':')
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = ':')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x,siny,label = 'sin(x)')
plt.plot(x,cosy,color = 'red',linestyle = ':',label = 'cos(x)')
plt.legend()
#加上图示
plt.show()
```

```python
plt.plot(x,siny,label = 'sin(x)')
plt.plot(x,cosy,color = 'red',linestyle = ':',label = 'cos(x)')
plt.legend()
plt.title('hello world')
plt.show()
```

### 散点图

```python
plt.scatter(x,siny)
plt.show()
```

```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
plt.scatter(x,y,alpha = 0.3)
plt.show()
```
